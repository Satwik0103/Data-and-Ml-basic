## Analysing the data

### Analysis of IPL Teams

Total no of teams

```python
teams=data1['winner'].value_counts()
len(teams)
```

```python
teams=data1['winner'].value_counts()
teams=teams.plot.barh(figsize=(15,8),color='yellow',fontsize=15)
plt.style.use('dark_background')
teams.set_title('Teams with most no of wins\n',fontsize=18,fontweight='bold')
teams.set_xlabel('No of matches won\n',fontsize=18,fontweight='bold')
teams.invert_yaxis()
for i in teams.patches:
    teams.text(i.get_width()+0.5,i.get_y()+.35,i.get_width(),fontsize=15,fontweight='bold',color='white')
```

```python
winning_teams = data1[['season','winner']]
winner_team ={}
for i in sorted(winning_teams.season.unique()):
    winner_team[i] = winning_teams[winning_teams.season ==i]['winner'].tail(1).values[0]
    
winner_IPL = pd.Series(winner_team)
winner_IPL = pd.DataFrame(winner_IPL, columns=['winner team'])
print("Season-wise Winner",winner_IPL)
```

```python
plt.figure(figsize=(16,8))
winner_IPL['winner team'].value_counts().plot(kind='bar', color='b')
plt.title('Winner of IPL across 11 seasons', fontsize=15)
plt.xlabel('Team', fontsize=15)
plt.ylabel('Number of seasons', fontsize=15)
```

### Analysis of TOSS

Decision taken by captain after winning the toss

```python
plt.figure(figsize=(16,8))
(data1['toss_decision'].value_counts()).plot(kind='pie',autopct = '%1.1f%%', colors=['r','orange'], fontsize=15)
plt.title('Decision taken by captain after winning toss', fontsize=15)
```

Individual team toss decision after winning toss

```python
plt.figure(figsize=(16,8))
sns.countplot(x='toss_winner',data = data1, hue='toss_decision', palette='rocket')
plt.title('Individual team toss decision after winning toss', fontsize=15)
plt.xlabel('Teams', fontsize=15)
plt.xticks(rotation=90)
plt.ylabel('Frequency', fontsize=15)
```

Matches won after batting first after winning the toss

```python
data1['win_by']=np.where(data1['win_by_runs']>0,'Bat first','Ball first')
```

```python
win=data1.win_by.value_counts()
labels=np.array(win.index)
sizes=win.values
colors=['red','blue']
plt.figure(figsize=(10,8))
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.title('Match win by',fontsize=20)
plt.axis('equal',fontsize=10)
plt.show()
```

For further analysis we need to merge the two datas

```python
data=pd.merge(data1,data2,left_on='id',right_on='match_id')
```

## Analysis of IPL Players

Top 10 players with most man of the match award

```python
plt.figure(figsize=(16,8))
sns.countplot(x='player_of_match',data = data1,order=data['player_of_match'].value_counts().head(10).index , palette='rainbow')
plt.title('Top 10 players with most man of the match ', fontsize=15)
plt.xlabel('Players', fontsize=15)
plt.xticks(rotation=90)
plt.ylabel('Frequency', fontsize=15)
```

Top 10 wicket taking bowlers in IPL

```python
data2.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(16,8), color='orange')
plt.title('Top 10 Wicket taking Bowlers in IPL', fontsize=15)
plt.xlabel('Bowlers', fontsize=15)
plt.xticks(rotation=90)
plt.ylabel('Total Wicket Taken', fontsize=15)
```

Top 10 batsmen in IPL

```python
#using merged dataset (mathces and deliveries)
data2.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(16,8), color='blue')
plt.title('Top 10 Batsmen in IPL', fontsize=15)
plt.xlabel('Batsmen', fontsize=15)
plt.xticks(rotation=90)
plt.ylabel('Total runs scored', fontsize=15)
```

## Analysis of the venue and result of match

Venue which hosted most number of IPL matches

```python
plt.figure(figsize=(16,8))
sns.countplot(x='venue',data=data1 ,order=data['venue'].value_counts().index)
plt.title('Venue which hosted most number of IPL matches', fontsize=15)
plt.xlabel('Venue', fontsize=15)
plt.xticks(rotation=90)
plt.ylabel('Number of matches', fontsize=15)
```

Season wise total no of matches played

```python
season_wise_matches=data1['season'].value_counts()
season_wise_matches=season_wise_matches.plot.barh(figsize=(15,8),color='crimson',fontsize=15)
plt.style.use('dark_background')
season_wise_matches.set_title('Season wise total no of Matches\n',fontsize=22,fontweight='bold')
season_wise_matches.set_xlabel('No of matches \n',fontsize=18,fontweight='bold')
season_wise_matches.set_xlabel('Season\n',fontsize=18,fontweight='bold')
season_wise_matches.invert_yaxis()
for i in season_wise_matches .patches:
    season_wise_matches.text(i.get_width()+0.5,i.get_y()+.35,i.get_width(),fontsize=15,fontweight='bold',color='white')
```

Match Results

```python
match_result=data1['result'].value_counts()
match_result=match_result.plot(kind='bar',figsize=(15,8),color='lightgreen',fontsize=15)
plt.style.use('dark_background')
match_result.set_title('Match Results\n',fontsize=18,fontweight='bold')
match_result.set_ylabel('Count of No of matches \n',fontsize=18,fontweight='bold')
totals=[]
for i in match_result.patches:
    totals.append(i.get_height())
total=sum(totals)
    
for i in match_result.patches:
    match_result.text(i.get_x()+0.10,i.get_height()+.01,str(np.round((i.get_height()/total)*100,2))+'%',fontsize=15,fontweight='bold',color='white')
```
